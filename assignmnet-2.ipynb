```python
## importing packages

import pandas as pd
import numpy as np
```

```python
# reading the data from csv files

bigkart_newsales_df = pd.read_csv('bigkart_newsales.csv')
bigkart_sales_df = pd.read_csv('bigkart_sales.csv')
```

### Question 1: 

Some of the orders are stored in another csv file named `bigkart_newsales`. Read the csv file, store it in a data frame and add it to the `bigkart_sales` data frame. Find the total sales value of the category 'Office Supplies' after combining the dataframes

```python
## combining / appending both bigkart_newsales_df & bigkart_sales_df

final_sales_df = pd.concat([bigkart_newsales_df, bigkart_sales_df]).reset_index(drop=True)
final_sales_df
```

```python
# subsetting 'Office Supplies' category and take the sume of Sales

office_sipplies_sales = final_sales_df[final_sales_df['Category']=='Office Supplies']['Sales'].sum()
print('Total Sales value of the category "Office Supplies" is ₹{}'.format(office_sipplies_sales))
```

### Question 2: 

There are some duplicate rows in the data frame. Drop these rows and calculate the total sales value of the category Office Supplies.

```python
# drop duplicated rows

final_sales_df = final_sales_df.drop_duplicates().reset_index(drop=True)
final_sales_df
```

```python
office_sipplies_sales = final_sales_df[final_sales_df['Category']=='Office Supplies']['Sales'].sum()
print('Total Sales value of the category "Office Supplies" after dropping duplicated rows is ₹{}'.format(office_sipplies_sales))
```

### Question 3:

Find the most profitable category and sub category combination based on the net profit.

```python
# groupby Category & Sub-Category columns and take sum of profit

cat_subcat_profit_gb = final_sales_df.groupby(['Category', 'Sub-Category']).agg({'Profit':'sum'}).reset_index()

# sorting the values based on total profit combined Category & Sub Category
most_profit_cat_subcat = cat_subcat_profit_gb.sort_values(by='Profit', ascending=False).reset_index(drop=True).head()
most_profit_cat_subcat
```

```python
print('{} - {} are the most profitable category and sub category combination with net profit of ₹{}, \nwhen compared to others'.format(most_profit_cat_subcat['Category'][0], most_profit_cat_subcat['Sub-Category'][0], most_profit_cat_subcat['Profit'][0]))
```

### Question 4: 

How many invalid order IDs are there in the data frame. An order id is of the form AZ-2011-Y where Y represents a whole number. A Order ID is said to be valid only if Y consists of 7 digits. Find the number of invalid order order IDs in the data frame.

```python
# split the order ID based on "-" (delimiter), which outputs list of strings, then index the last string from the list and 
# take length of string to validate if the oder id is valid or invalid
final_sales_df['order_id_val_check'] = final_sales_df['Order ID'].apply(lambda x: x.split("-")[-1]).str.len()

# create datafatme, take count of values based on newly created column 'order_id_val_check' and renaming them 
coun_val_inval_df = final_sales_df['order_id_val_check'].value_counts().reset_index().rename(columns = {'index': 'len_oder_id', 'order_id_val_check':'count_val_inval'})
coun_val_inval_df
```

```python
print('{} Order IDs are invalid of {}' .format(coun_val_inval_df[coun_val_inval_df['len_oder_id']==6]['count_val_inval'][1], len(final_sales_df)))
```

### Question 6: 

Among the orders with sales>250 and profit>50, find the product name of the fourth highest order based on sales value.

```python
# subset the dataframe with Sales>250 & Profit>50, sort descending based on Sales 

prod_four_high = final_sales_df[(final_sales_df['Sales']>250)&(final_sales_df['Profit']>50)].sort_values(by='Sales', ascending=False).reset_index(drop=True)

print('Among the orders with sales>250 and profit>50, "\033[1m{}\033[0m" is fourth highest order based on sales value.' .format(prod_four_high.loc[3, 'Product Name']))
```

### Question 7: 

Remove the orders with negative profit by dropping the corresponding rows with negative `Profit`. Find the product that makes the lowest profit per Quantity in the Technology category.

```python
# remove the profits with negative value, subset Technology Category and order values ascending based on Profit

non_neg_prof_tech_cat_df = final_sales_df[(final_sales_df['Profit']>=0)&(final_sales_df['Category']=='Technology')].sort_values('Profit').reset_index(drop=True)
non_neg_prof_tech_cat_df
```

```python
print('\033[1m{}\033[0m is the product with least Profit value of \033[1m₹{}\033[0m amoung Category Technology'.format(non_neg_prof_tech_cat_df.loc[0, 'Product Name'], non_neg_prof_tech_cat_df.loc[0, 'Profit']))
```
